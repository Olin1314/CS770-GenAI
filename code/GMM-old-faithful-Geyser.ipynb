# Guassian Mixture Models: Old Faithful Geyser Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
df = pd.read_csv('http://www.biostat.jhsph.edu/~rpeng/useRbook/faithful.csv')
```

```python
X = df.values
# scale X
X = (X - X.mean()) / X.std()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1])
```

```python
def visualize(faith_gmm,X):
    # predict labels
    predicted_y = faith_gmm.predict(X)
    # plot
    plt.figure(figsize=(8,8)) # set size
    # plot old faithful data
    plt.scatter(X[:, 0], X[:, 1], alpha = 0.8, c=predicted_y, cmap='viridis')
    # plot gmm centroid
    plt.plot(faith_gmm.means_[:, 0], faith_gmm.means_[:, 1], color = 'red', marker= 'X', markersize = 18, linewidth = 0);
    # title
    plt.title('Old Faithful: GMM Clusters');
```

```python
faith_gmm = GaussianMixture(n_components=1, 
                            covariance_type='full', 
                            n_init=1, 
                            random_state=1,max_iter=1000).fit(X)
visualize(faith_gmm,X)
```

```python
faith_gmm = GaussianMixture(n_components=2, 
                            covariance_type='full', 
                            n_init=1, 
                            random_state=1,max_iter=2).fit(X)

print(faith_gmm.means_)

visualize(faith_gmm,X)
```

```python
faith_gmm = GaussianMixture(n_components=2, 
                            covariance_type='full', 
                            n_init=1, 
                            random_state=1,max_iter=100).fit(X)
print(faith_gmm.means_)

visualize(faith_gmm,X)
```

```python
faith_gmm = GaussianMixture(n_components=2, 
                            covariance_type='full', 
                            n_init=1, 
                            random_state=1,max_iter=1000).fit(X)
print(faith_gmm.means_)

visualize(faith_gmm,X)
```
